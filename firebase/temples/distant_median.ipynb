## 情報収集先URL
- おへんろずかん：https://sanpo.xyz/distance/
- 四国巡礼コース：http://walking.ponyo.jp/course.php?c=4
- 四国八十八ヶ所お遍路：http://ohenro-88.com/arukihenro/howfar.html

#### おへんろずかん

```python
# ページの情報を取得
url = "https://sanpo.xyz/distance/"
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# <td class="has-text-align-center">に距離の情報がある
table_data = [data.text for data in soup.find_all("td",class_="has-text-align-center")]
```

```python
# 距離の情報のみを抽出する
distant_list1 = []
for td in table_data:
    if 'km' in td or 'mk' in td:
        distant = td.split("km")[0].replace(",",".") 
        distant = distant.split("mk")[0].replace(",",".") 
        distant_list1.append(float(distant))
```

```python
len(distant_list1)
```

#### 四国巡礼コース

```python
# ページの情報を取得
url = "http://walking.ponyo.jp/course.php?c=4"
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# <td class="has-text-align-center">に距離の情報がある
table_data = [data.text for data in soup.find_all("td")]
```

```python
# 距離のデータのみを取得する
distant_list2 = []
for td in  table_data:
        if 'km' in td:
            distant = td.split("km")[0].replace(",","")
            distant_list2.append(float(distant))
```

```python
# 区間距離である奇数番目を取得
distant_list2 = distant_list2[1::2][1:]
```

```python
# １に戻る距離がなかったのでデータなしの"-"を追加する
distant_list2.append('-')
```

#### 四国八十八ヶ所お遍路

```python
# ページの情報を取得
url = "http://ohenro-88.com/arukihenro/howfar.html"
res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# <td class="has-text-align-center">に距離の情報がある
table_data = [data.text for data in soup.find_all("td")]
```

```python
# 距離のデータのみを取得する
distant_list3 = []
for td in  table_data:
        if 'km' in td:
            distant = td.split("km")[0]
            distant_list3.append(float(distant))
```

```python
len(distant_list3)
```

```python
# データをまとめる
df = pd.DataFrame({"おへんろずかん":distant_list1,
                                   "四国巡礼コース":distant_list2,
                                   "四国八十八ヶ所お遍路":distant_list3})
```

```python
df
```

```python
med = df.median(axis=1).to_list()
```

```python
med = [int(d*1000) for d in med]
```

```python
med
```

```python
# データの読み込み
df_main = pd.read_csv("temple_info.csv",encoding="utf-8")
df_main["徒歩距離(m)"] = med
df_main.to_csv("temple_info_update.csv",encoding="utf-8")
```
